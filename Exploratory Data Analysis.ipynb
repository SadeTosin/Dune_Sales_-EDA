### Data/time Analysis

```python
df.head(3)
```

```python
# Converting the pandas column into pandas date and time object
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.info()
```

```python
# extract the Year, Month and Quarter
```

```python
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Month_Name"] = df["Date"].dt.month_name()
df["Quarter"] = df["Date"].dt.quarter
```

```python
df.head(3)
```

```python
# customer age group
```

```python
def age_group(x):
    if x <= 25:
        return "<= 25 Young Adult"
    elif x <= 40:
        return "<= 40 Adult"
    elif x <= 50:
        return "<= 50 Old Adult"
    else:
        return ">= 51 Elder"

# Apply functipon to the data
df["Age_Group"] = df["Customer_Age"].apply(age_group)
```

```python
df.head(3)
```

```python
# Cost, Revenue and Profit Calculation
```

```python
df["Cost"] = df["Quantity"] * df["Unit_Cost"]
df["Revenue"] = df["Quantity"] * df["Unit_Price"]
df["Profit"] = df["Revenue"] - df["Cost"]

df.head(3)
```

```python
# determine profit or loss
```

```python
def pro_los(x):
    if x >= 0:
        return "profit"
    else:
        return "loss"
#Applying the function to DataFrame

df["Prof_OR_Loss"] = df["Profit"].apply(pro_los)

df.head(3)
```
